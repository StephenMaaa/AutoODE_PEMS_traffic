### Data Processing 

```python
# script to extract Total_flow and Avg_speed for each sensor each day 
# PEMS data in 04/01 - 09/30 in 2021 03/16 - 09/16

import itertools 
import torch 

datas = []
for filename in os.listdir("caltrans/station_5min/2021/d11_plaintxt/"): 
    print(filename)
#     if int(filename[-9:-7]) < 9: 
#         continue
#     if "d11_text_station_5min_2021_04_04.txt" in filename: 
#         continue
#     if "d11_text_station_5min_2021_06_17.txt" in filename: 
#         continue
#     if "d11_text_station_5min_2021_06_18.txt" in filename: 
#         continue
    filedata = pd.read_csv("caltrans/station_5min/2021/d11_plaintxt/" + filename, sep=",", usecols=range(12), names = col)       
    
    # filter out selected sensor ids
    filedata = filedata[(filedata["Freeway"] == 5) & (filedata["Direction"] == "S") & (filedata["Type"] == "ML")]
    filedata = filedata[filedata["Station"].isin(sorted_sensors)]
    
    # sort dataframe given ids and timestamps
    if "d11_text_station_5min_2021_09_14.txt" in filename: 
        time = filedata["Timestamp"].unique().tolist()[:288] 
        filedata = filedata[filedata["Timestamp"].isin(time)]
    else: 
        time = filedata["Timestamp"].unique().tolist() 
        
    pair = list(itertools.product(time, sorted_sensors))
    filedata["Station"] = filedata["Station"].astype("category")
    filedata["Timestamp"] = filedata["Timestamp"].astype("category")
    filedata["Station"].cat.set_categories(sorted_sensors, inplace=True)
    filedata["Timestamp"].cat.set_categories(time, inplace=True)
    filedata = filedata.sort_values(["Station"])
    for i in range(1, len(sorted_sensors) + 1): 
        filedata.iloc[288 * (i - 1) : 288 * i] = filedata.iloc[288 * (i - 1) : 288 * i].sort_values(["Timestamp"])
        
    data = filedata[["Flow", "Avg Speed"]].to_numpy()
    data = data.reshape(len(sorted_sensors), -1)
    data = data.reshape(len(sorted_sensors), 288, 2) 
    data = torch.tensor(data).permute(2, 1, 0)
    print(data.isnan().any())
    datas.append(data)
#     print(data.shape)
```
