```python
import gzip
import shutil 
```

```python
import os

count = 0
for filename in os.listdir("caltrans/station_5min/2021/d4/"): 
    plaintxt, file_type = filename.split(".gz")
    print(plaintxt)
    with gzip.open("caltrans/station_5min/2021/d4/" + filename, 'rb') as f_in:
        with open("caltrans/station_5min/2021/d4_plaintxt/" + plaintxt, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
            count += 1
```
