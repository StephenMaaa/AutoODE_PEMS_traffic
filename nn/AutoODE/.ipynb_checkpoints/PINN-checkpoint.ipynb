```python
import deepxde as dde
import numpy as np
# Backend tensorflow.compat.v1 or tensorflow
from deepxde.backend import tf
# Backend pytorch
# import torch
```

```python
import torch
```

```python
# first 68 timesteps for training, last 3 for testing

def generate_data(): 
    y_exact=torch.load("5S_191111_3cmp_torch.pt")  ## 3 x 71 x 31
#     x = np.arange(0, 120, 1)[:, None]
    t = np.arange(0, 68, 1)[:, None]
    xi = np.array([0, 4, 12, 22, 32, 39, 46, 50, 54, 57, 59, 63, 66, 69, 71, 74, 
                   76, 81, 84, 86, 89, 91, 95, 96, 98, 100, 102, 105, 108, 110, 112])[:, None]
    X, T = np.meshgrid(xi, t) 
    X = np.vstack((np.ravel(X), np.ravel(T))).T
    k = y_exact[0, :68, :].flatten()[:, None]
    q = y_exact[1, :68, :].flatten()[:, None]
    u = y_exact[2, :68, :].flatten()[:, None]
    
    # test data 
    t_t = np.arange(68, 71, 1)[:, None]
    X_t, T_t = np.meshgrid(xi, t_t) 
    X_t = np.vstack((np.ravel(X_t), np.ravel(T_t))).T
    k_t = y_exact[0, 68:, :].flatten()[:, None]
    q_t = y_exact[1, 68:, :].flatten()[:, None]
    u_t = y_exact[2, 68:, :].flatten()[:, None]
    return (X, k, q, u), (X_t, k_t, q_t, u_t)
```

```python
(X, k, q, u), (X_t, k_t, q_t, u_t) = generate_data()
```

```python
def pde(x, y):
    dq_x = dde.grad.jacobian(y, x, i=1, j=0)
    dk_t = dde.grad.jacobian(y, x, i=0, j=1)
    return dq_x + dk_t
```

```python
# define geometry and time domain 
geom = dde.geometry.Interval(0, 120)
timedomain = dde.geometry.TimeDomain(0, 68)
geomtime = dde.geometry.GeometryXTime(geom, timedomain)
```

```python
# observed training points (using training data) 
observe_y1 = dde.PointSetBC(X, k, component=0)
observe_y2 = dde.PointSetBC(X, q, component=1)
observe_y3 = dde.PointSetBC(X, u, component=2)
```

```python
# initial condition and boundary condition 

#nx=350  # 
y_exact=torch.load("5S_191111_3cmp_torch.pt")
nx=120
#kj needs to be larger than k for the solution to be stable 
kj = np.ones((nx,), dtype=float) * 0.2
# kj = (kj - y_min[0].numpy()) / (y_max[0].numpy() - y_min[0].numpy()) # normalize

#characteristic velocity (m/s), corresponds to roughly 120 km/h
vf = np.ones((nx,), dtype=float) * 33
# vf = (vf - y_min[2].numpy()) / (y_max[2].numpy() - y_min[2].numpy()) # normalize

dx=300.
dt=6.
#need an output every 5 mins (300 s), so tskip = 3 with dt = 3s
tskip=50
#nt=int(3600*6/6 - 50)
#nt=7099 #6 hours (times 3600 s/hour divided by dt=3s)
nto=(y_exact.shape[1]+1) 
#nt=int(3600*nto/12/6/dt - tskip)
dtobs=300
nt=int((dtobs*nto)/dt - tskip)
print(dt,nt)
```

```python
xi = [0, 4, 12, 22, 32, 39, 46, 50, 54, 57, 59, 63, 66, 69, 71, 74, 
      76, 81, 84, 86, 89, 91, 95, 96, 98, 100, 102, 105, 108, 110, 112]
```

```python
from scipy.interpolate import interp1d
print(xi)
def interpolate_initial(var, t0=0):
    IM=interp1d(np.array(xi)*dx, var[t0,:].numpy(), bounds_error=False, 
                fill_value=(var[t0,0], var[t0,-1]), kind='linear')
    
    x=np.linspace(0, (nx-1)*dx, nx)
    return IM(x)
```

```python
k_initial={}
cmps = ["k","q","u"]
for n,c in enumerate(cmps):
    k_initial[c]=interpolate_initial(y_exact[n,:])
    print(k_initial[c].max())
    print(k_initial[c].shape)
```

```python
ti=np.arange(0, (nt-1)*dt, tskip*dt )
print(len(ti), y_exact.shape[0])
t=np.linspace(0, (nt-1)*dt, nt)
#print(len(ti), len(y_exact))
#print(t)

k_boundary={}

for n,c in enumerate(cmps):
    IMb=interp1d(np.array(ti), y_exact[n,:,0].numpy(), bounds_error=False,
             fill_value=(y_exact[n,0,0], y_exact[n,-1,0]),
             kind='linear')
    k_boundary[c]=IMb(t)
    print(k_boundary[c].shape, k_boundary[c].max())
```

```python
initial = np.stack((k_initial["k"], k_initial["q"], k_initial["u"]))
boundary = np.stack((k_boundary["k"][::10][:-15], k_boundary["q"][::10][:-15], k_boundary["u"][::10][:-15]))
```

```python
boundary
```

```python
def generate_initial(initial): 
#     y_exact=torch.load("5S_191111_3cmp_torch.pt")  ## 3 x 71 x 31
    x = np.arange(0, 120, 1)[:, None]
    t = np.arange(0, 1, 1)[:, None]
    X, T = np.meshgrid(x, t) 
    X = np.vstack((np.ravel(X), np.ravel(T))).T
    k = initial[0, :].flatten()[:, None]
    q = initial[1, :].flatten()[:, None]
    u = initial[2, :].flatten()[:, None]
    return X, k, q, u
```

```python
def generate_boundary(boundary): 
#     y_exact=torch.load("5S_191111_3cmp_torch.pt")  ## 3 x 71 x 31
    x = np.arange(0, 1, 1)[:, None]
    t = np.arange(0, 68, 0.2)[:, None]
    X, T = np.meshgrid(x, t) 
    X = np.vstack((np.ravel(X), np.ravel(T))).T
    k = boundary[0, :].flatten()[:, None]
    q = boundary[1, :].flatten()[:, None]
    u = boundary[2, :].flatten()[:, None]
    return X, k, q, u
```

```python
X_i, k_i, q_i, u_i = generate_initial(initial)
X_b, k_b, q_b, u_b = generate_boundary(boundary)
```

```python
initial_c1 = dde.PointSetBC(X_i, k_i, component=0)
initial_c2 = dde.PointSetBC(X_i, q_i, component=1)
initial_c3 = dde.PointSetBC(X_i, u_i, component=2)
```

```python
boundary_c1 = dde.PointSetBC(X_b, k_b, component=0)
boundary_c2 = dde.PointSetBC(X_b, q_b, component=1)
boundary_c3 = dde.PointSetBC(X_b, u_b, component=2)
```

```python
data = dde.data.TimePDE(
        geomtime,
        pde,
        [initial_c1, initial_c2, initial_c3, boundary_c1, boundary_c2, boundary_c3, observe_y1, observe_y2, observe_y3],
        num_domain=50000,
        num_boundary=50000, 
        #num_initial=150,
        train_distribution="uniform",
        anchors=X,
        num_test=5000,
    )
```

```python
# create the train and the model 
net = dde.maps.FNN([2] + [64] * 4 + [3], "tanh", "Glorot uniform")
model = dde.Model(data, net)
model.compile("adam", lr=0.001)
```

```python
# train
losshistory, train_state = model.train(epochs=50000)
# results
dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```

### Evaluation 

```python
# training error

(X, k, q, u), (X_t, k_t, q_t, u_t) = generate_data()
y_pred = model.predict(X)
f = model.predict(X, operator=pde)
print("Mean residual:", np.mean(np.absolute(f)))
y_true=np.concatenate((k, q, u), axis=1)
print("L2 relative error:", dde.metrics.l2_relative_error( y_true, y_pred)) 
print("Mean square error:", dde.metrics.mean_squared_error( y_true, y_pred))
```

```python
# test error

y_pred = model.predict(X_t)
f = model.predict(X_t, operator=pde)
print("Mean residual:", np.mean(np.absolute(f)))
y_true=np.concatenate((k_t, q_t, u_t), axis=1)
print("L2 relative error:", dde.metrics.l2_relative_error( y_true, y_pred)) 
print("Mean square error:", dde.metrics.mean_squared_error( y_true, y_pred))
```
