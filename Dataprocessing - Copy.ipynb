```python
def interpolate(var, ti, t): 
    IM_q=interp1d(np.array(ti), var[:, 0], bounds_error=False,
             fill_value=(var[0, 0], var[-1, 0]), kind='linear')
    IM_u=interp1d(np.array(ti), var[:, 1], bounds_error=False,
             fill_value=(var[0, 1], var[-1, 1]), kind='linear')
    u = IM_u(t)
    q = IM_q(t)
    boundary = np.stack((q, u), axis = 1) 
#     print(boundary.shape)
    return torch.tensor(boundary)
```
